## Dataclean

```python
import pandas as pd
import datetime as dt
now=dt.datetime.now()
today=now.strftime('%Y_%m_%d')
Y=int(now.strftime('%Y'))
M=int(now.strftime('%m'))
dfgl=pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
dfgl_date=dfgl['date'].str.split('-',5, True)
dfgl['Y']=dfgl_date[0]
dfgl['M']=dfgl_date[1]
dfgl['D']=dfgl_date[2]

dfgl['Y']=dfgl['Y'].astype('int')
dfgl['M']=dfgl['M'].astype('int')
dfgl['D']=dfgl['D'].astype('int')

Mdata_gl=dfgl.groupby(['location','Y','M'])['new_cases'].sum().reset_index()
Mdata_gl['time']=pd.to_datetime(Mdata_gl['Y']+'-'+Mdata_gl['M']+'-'+'1', format='%Y_%m_%d')
Mdata_gl['time_str']=Mdata_gl['Y']+'-'+Mdata_gl['M']+'-'+'1'
Mdata_gl['Y']=Mdata_gl['Y'].astype('int')
Mdata_gl['M']=Mdata_gl['M'].astype('int')
Mdata_gl
```

## trend_analys_MLY

```python
maskY=Mdata_gl['Y']!=Y
maskM=Mdata_gl['M']!=M
Mdata_gl1=Mdata_gl[maskY|maskM]
maskunC1=Mdata_gl1['location']!='World'
maskunC2=Mdata_gl1['location']!='Asia'
maskunC3=Mdata_gl1['location']!='Europe'
maskunC4=Mdata_gl1['location']!='Africa'
maskunC5=Mdata_gl1['location']!='North America'
maskunC6=Mdata_gl1['location']!='South America'
maskunC7=Mdata_gl1['location']!='Oceania'
Mdata_contry=Mdata_gl1[maskunC1&maskunC2&maskunC3&maskunC4&maskunC5&maskunC6&maskunC7]
Mdata_contry.to_excel(r'D:\jupter_data\covid19\country_covid19_Mdata.xlsx')
##
maskunCt1=Mdata_gl1['location']=='World'
maskunCt2=Mdata_gl1['location']=='Asia'
maskunCt3=Mdata_gl1['location']=='Europe'
maskunCt4=Mdata_gl1['location']=='Africa'
maskunCt5=Mdata_gl1['location']=='North America'
maskunCt6=Mdata_gl1['location']=='South America'
maskunCt7=Mdata_gl1['location']=='Oceania'
Mdata_continent=Mdata_gl1[(maskunCt1|maskunCt2|maskunCt3|maskunCt4|maskunCt5|maskunCt6|maskunCt7)]
Mdata_continent.to_excel(r'D:\jupter_data\covid19\Continent_covid19_Mdata.xlsx')
```

## map_analys

```python
todayS=dt.date.today()
towday = dt.timedelta(days=2)
Byesterday=todayS-towday
Byesterday_str=Byesterday.strftime('%Y-%m-%d')
dfgl_yes=dfgl[dfgl['date']==Byesterday_str]
maskunC1=dfgl_yes['location']!='World'
maskunC2=dfgl_yes['location']!='Asia'
maskunC3=dfgl_yes['location']!='Europe'
maskunC4=dfgl_yes['location']!='Africa'
maskunC5=dfgl_yes['location']!='North America'
maskunC6=dfgl_yes['location']!='South America'
maskunC7=dfgl_yes['location']!='Oceania'
dfgl_yes1=dfgl_yes[(maskunC1&maskunC2&maskunC3&maskunC4&maskunC5&maskunC6&maskunC7)]
dfgl_yes1.to_excel(r'D:\jupter_data\covid19\country_covid19_onedaydata.xlsx')
```
