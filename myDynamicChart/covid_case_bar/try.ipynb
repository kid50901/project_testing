```python
import pandas as pd
df=pd.read_csv('https://od.cdc.gov.tw/eic/Age_County_Gender_19Cov.csv')
df
```

```python
dfG=df.groupby(['發病年份','發病月份','縣市'])['確定病例數'].sum().reset_index()
mask1=dfG['發病年份']==2021
mask2=dfG['發病月份']==5
dfGthis=dfG[(mask1&mask2)]
dfGthis.groupby(['發病年份','發病月份'])['確定病例數'].sum().reset_index()
```

```python
import pandas as pd
dfgl=pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
```
