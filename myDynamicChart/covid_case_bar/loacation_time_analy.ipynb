## Dataclean

```python
import pandas as pd
import datetime as dt
now=dt.datetime.now()
today=now.strftime('%Y_%m_%d')
Y=int(now.strftime('%Y'))
M=int(now.strftime('%m'))
dfgl=pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv')
dfgl_date=dfgl['date'].str.split('-',5, True)
dfgl['Y']=dfgl_date[0]
dfgl['M']=dfgl_date[1]
dfgl['D']=dfgl_date[2]

#dfgl['Y']=dfgl['Y'].astype('int')
#dfgl['M']=dfgl['M'].astype('int')
#dfgl['D']=dfgl['D'].astype('int')

Mdata_gl=dfgl.groupby(['location','Y','M'])['new_cases'].sum().reset_index()
Mdata_gl['time']=pd.to_datetime(Mdata_gl['Y']+'_'+Mdata_gl['M']+'_'+'1', format='%Y_%m_%d')
Mdata_gl['time_str']=Mdata_gl['Y']+'/'+Mdata_gl['M']
Mdata_gl['Y']=Mdata_gl['Y'].astype('int')
Mdata_gl['M']=Mdata_gl['M'].astype('int')
Mdata_gl
```
